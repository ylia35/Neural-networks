## Задача 1.2

```python
X1 = X.ravel()
n = len(X1)
Sx = np.sum(X1)
Sy = np.sum(y)
Sxx = np.sum(X1**2)
Sxy = np.sum(X1 * y)
SX = Sx**2

print(n, ' ', Sx, ' ', Sy, ' ', Sxx, ' ', Sxy, ' ', SX)
```

```python
w1 = (n * Sxy - Sx * Sy)/(n * Sxx - SX)

print(w1)
```

```python
b1 = (Sy * Sxx - Sx * Sxy)/(n * Sxx - SX)

print(b1)
```

```python
xmin, xmax = X1.min(), X1.max()
x_line = np.array([xmin, xmax])

y_theory = w * x_line + b
y_opt = w1 * x_line + b1

plt.figure()
plt.scatter(X1, y, s=10, label="Данные", color=("green"))
plt.plot(x_line, y_theory, linewidth=2, label="Теоретическая линия", color="red")
plt.plot(x_line, y_opt, linewidth=2, linestyle="--", label="Оптимальная линия", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Точки, теоретическая и оптимальная линии")
plt.show()
```

Реализуйте предложенный ниже шаблон класса линейной регресиии numpyLinearRegression для произвольного количества признаков. Используйте формулы из лекции и numpy.linalg.

Проверьте работу класса на вашем датасете, убедитеcь, что оптимальные параметры  w^  и  b^  получаются те же самые (или разберитесь, почему это не так). Сравните результаты с sklearn.linear_model.LinearRegression.

[4 балла] добавляется, если решение запускается без ошибки выполнения на ваших данных

[1 балл] добавляется, если реализовано сравнение вашего ответа с ответом для sklearn.linear_model.LinearRegression

[4 балла] добавляется, если решение для вашей модели на ваших данных сходится с ответом sklearn.linear_model.LinearRegression

[5 баллов] добавляется, если решение полностью векторизовано (нет циклов)

[5 баллов] добавляется, если в решении обрабатывается случай, когда не существует обратная матрица, выводится соответствующее предупреждение и вычисления прерываются (либо продолжаются иначе на ваше усмотрение)

```python
class numpyLinearRegression:
    def __init__(self):
        self.coef_ = None # массив, который будет хранить веса при признаках X
        self.intercept_ = None # свободный параметр

    def fit(self, X: np.ndarray, y: np.ndarray):
        """
        X : (n_samples, n_features)
        y : (n_samples,)
        """
        X_ext = np.column_stack([X, np.ones(X.shape[0])])
        try:
            theta, residuals, rank, s = np.linalg.lstsq(X_ext, y, rcond=None)
        except np.linalg.LinAlgError:
            print("Решение может быть не единственное")
            theta = np.linalg.pinv(X_ext) @ y

        self.coef_ = theta[:-1] #все элементы кроме последнего - веса признаков
        self.intercept_ = theta[-1] # последний элемент - свободный член
        return self

    def predict(self, X: np.ndarray):
        """X : (n_samples, n_features) -> y_pred : (n_samples,)"""
        if self.coef_ is None or self.intercept_ is None:
            raise ValueError("Модель не обучена")
        return X @ self.coef_ + self.intercept_
```

```python
model = numpyLinearRegression().fit(X, y)

print("numpyLinearRegression:")
print("w =", model.coef_[0])
print("b =", model.intercept_)

sk = LinearRegression().fit(X, y)
print("\nsklearn:")
print("w =", sk.coef_[0])
print("b =", sk.intercept_)
```
